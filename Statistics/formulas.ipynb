# 2-Mode

```python
data=np.array([9,1,3,6,10,3,7])
print("The mode is ",statistics.mode(data))
```

# 3-median

```python
data=np.array([9,1,3,6,10,3,7])
print("The mode is:",statistics.median(data))
```

# 4-Variance

```python
data=np.array([9,1,3,6,10,3,7])
print("The Variance of given data is:",statistics.variance(data))
```

```python
import statistics
import numpy as np

# Example data
data1 = np.array([3,8,6,10,12,9,11,10,12,7])  # or however your data is defined

# Convert numpy array to list of floats
data1 = data1.astype(float).tolist()
# Now calculate standard deviation
print("The standard deviation of given data is:", statistics.stdev(data1))
```

```python
x=np.array([2,4,13,9])
print("The range is :",max(x)-min(x))
```

```python
x=np.array([2,4,13,9])
print("The midrange is :",(max(x)+min(x))/2)
```
